### Ejemplo con Matplotlib

```python
df_csv.columns
```

```python
# Calcular las ventas totales por -branch-
ventas_agrupadas = df_csv.groupby('Branch')["Sales"]
```

```python
ventas_agrupadas.plot(kind='bar', color='steelblue')
plt.title('Ventas Totales')
plt.ylabel('Monto en $$$')
plt.show()
```
